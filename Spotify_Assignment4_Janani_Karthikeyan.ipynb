# üåç **Task 4Ô∏è‚É£: Geospatial Visualization - Choropleth Map of Spotify Streams**

### **Objective**  
Create a **choropleth map** to visualize the distribution of Spotify streams across different countries, helping identify regional streaming trends.

---

### **What You‚Äôll Do**  
1. **Understand Geospatial Data & Visualization**:  
   - Geospatial data represents information tied to locations on Earth.  
   - Choropleth maps use color shading to represent numerical values across geographic regions.  
   - üì∫ Watch this [video](https://www.youtube.com/watch?v=ZS7v_8HhmG8) to learn more.

2. **Prepare the Data**:  
   - Extract **country-wise** streaming data from the dataset.  
   - Ensure country names match standardized geographic codes for mapping.

3. **Generate the Choropleth Map**:  
   - Use `plotly.express.choropleth()` to create an interactive visualization.  
   - Customize the map with labels, colors, and tooltips for clarity.

4. **Analyze Insights**:  
   - Identify countries with **highest and lowest** Spotify streams.  
   - Observe **regional streaming trends** and discuss possible reasons behind them.


```python
# ==== Task 4: Geospatial Visualization - Choropleth Map ====
# üöÄ Goal: Create a choropleth map to visualize the distribution of Spotify streams across countries.

# 1Ô∏è‚É£ Prepare the Data
# Extract country-wise streaming data and ensure country names are standardized.
# -------- Your code starts here --------
def get_country_alpha3(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3
    except LookupError:
        return None

# Apply the function and clean the data
df_cleaned['country_alpha3'] = df_cleaned['country'].apply(get_country_alpha3)
df_cleaned = df_cleaned.dropna(subset=['country_alpha3'])  # Remove unmatched countries

# Group data by country to get total streams per country
df_country_streams = df_cleaned.groupby(['country', 'country_alpha3'])['popularity'].sum().reset_index()

# -------- Your code ends here --------

# 2Ô∏è‚É£ Create the Choropleth Map
# Use plotly.express.choropleth() to visualize the data on a world map.
# -------- Your code starts here --------
fig = px.choropleth(df_country_streams,
                    locations="country_alpha3",
                    locationmode="ISO-3",
                    color="popularity",  # Total streams per country
                    title="Spotify Streams by Country",
                    color_continuous_scale="viridis",
                    hover_name="country")

# -------- Your code ends here --------

# 3Ô∏è‚É£ Customize and Display the Map
# Add labels, colors, and tooltips for better readability.
# -------- Your code starts here --------
fig.update_geos(projection_type="natural earth")
fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0})  # Remove extra whitespace
fig.show()

# -------- Your code ends here --------

# 4Ô∏è‚É£ Analyze Insights
# Identify top and bottom streaming countries and document key observations.
# -------- Your code starts here --------
top_countries = df_country_streams.nlargest(5, 'popularity')[['country', 'popularity']]
bottom_countries = df_country_streams.nsmallest(5, 'popularity')[['country', 'popularity']]

print("Top 5 Countries by Spotify Streams:")
print(top_countries.to_string(index=False))

print("\nBottom 5 Countries by Spotify Streams:")
print(bottom_countries.to_string(index=False))

# -------- Your code ends here --------
```

**Insights**

1. **Australia (AU)** leads in Spotify streams, surpassing traditional markets like the **US** and **Canada (CA)**.  
2. **New Zealand (NZ)** and the **United Arab Emirates (AE)** show high streaming numbers despite smaller populations.  
3. The **United States (US)** ranks fifth, which is unexpectedly low given its large user base.  
4. **Iceland (IS)**, **Egypt (EG)**, and **Morocco (MA)** have the lowest streams, likely due to smaller populations or regional platform preferences.  
5. Eastern European countries like **Belarus (BY)** and **Bulgaria (BG)** also rank low, hinting at differing music consumption habits.
